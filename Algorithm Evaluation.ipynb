```python
import pandas as pd
import numpy as np 
# numpy..numerical python deals with multidimensional array...matrices
import math
```

```python
Smartphones=pd.read_csv('updated_smartphone.csv')
Ratings=pd.read_csv('updated_rating.csv')
Mean= Ratings.groupby(['user_id'], as_index = False, sort = False).mean().rename(columns = {'ratings': 'rating_mean'})[['user_id','rating_mean']]
Ratings= pd.merge(Ratings, Mean, on ='user_id')
Ratings['rating_adjusted']=Ratings['ratings']-Ratings['rating_mean']
Ratings
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(Ratings, test_size=0.20) 
test_data
```

```python
test_rating_actual=test_data.loc[:,['user_id','smartphone_id','ratings']]
test_rating_actual
```

```python
train_data
```

```python
#make pivot table ..train data
train_matrix=pd.pivot_table(train_data,index='user_id',columns='smartphone_id',values='ratings') #value with actual ratings
train_matrix
```

```python
#make pivot table...test data
test_matrix=pd.pivot_table(test_data,index='user_id',columns='smartphone_id',values='ratings') #value with actual ratings
test_matrix
```

```python
#replacing values with adjusted ratings(normalization of ratings for cosine similarity)
final=pd.pivot_table(train_data,index='user_id', columns='smartphone_id', values='rating_adjusted')
final
```

```python
#if any NaN values is there then replace that with the average of users
# Replacing NaN by user Average
final_user = final.apply(lambda row: row.fillna(row.mean()), axis=1)
#matrix with NaN value is a sparse matrix
final_user
```

```python
# Replacing NaN by Gadget Average
final_gadget = final.fillna(final.mean(axis=0))
final_gadget
```

```python
#USING USER AVERAGE

from sklearn.metrics.pairwise import cosine_similarity
b = cosine_similarity(final_user)
np.fill_diagonal(b, 0 )
similarity_with_user = pd.DataFrame(b,index=final_user.index)
similarity_with_user.columns=final_user.index
similarity_with_user
```

```python
#USING GADGET AVERAGE

cosine = cosine_similarity(final_gadget)
np.fill_diagonal(cosine, 0 )
similarity_with_gadget = pd.DataFrame(cosine,index=final_gadget.index)
similarity_with_gadget.columns=final_user.index
similarity_with_gadget.head()
```

```python

def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1) #axis 0 for index 1 for columns
    return df
```

```python
# top 3 neighbours for each user
sim_user_8_u = find_n_neighbours(similarity_with_user,8)
sim_user_8_u.head()
```

```python
#list of items rated by every users
train_data = train_data.astype({"smartphone_id": str})
gadget_user = train_data.groupby(by = 'user_id')['smartphone_id'].apply(lambda x:','.join(x))
gadget_user
```

```python
gadget_seen_by_user=train_matrix.columns[train_matrix[train_matrix.index==4].notna().any()].tolist()
gadget_seen_by_user   
```

```python
users = list(set(train_data['user_id'].tolist()))
users
```

```python
def User_item_score2(user_list):
    final_list=[]
    for user in user_list:
        
        gadget_seen_by_user=train_matrix.columns[train_matrix[train_matrix.index==user].notna().any()].tolist()
        a=sim_user_8_u[sim_user_8_u.index==user].values #similar users to a particular user
        b=a.squeeze().tolist()
        c=gadget_user[gadget_user.index.isin(b)]
        l=','.join(c.values)
        gadget_seen_by_similar_users=l.split(',')
        gadget_under_considiration=list(set(gadget_seen_by_similar_users)-set(list(map(str,gadget_seen_by_user))))
        gadget_under_considiration=list(map(int, gadget_under_considiration))
        score= {'user_id':[],'item_id':[], 'Rating':[]}

        for item in gadget_under_considiration:
            d=final_user.loc[:,item] #sabai uer le tyo item lai deko ratings
            e=d[d.index.isin(b)] #similar users le deko ratings extract gariyo
            f=e[e.notnull()]
            avg_user=train_data.loc[train_data['user_id']==user,'rating_mean'].values[0]
            index = f.index.values.squeeze().tolist() #neighbour users
            corr=similarity_with_user.loc[user,index]
            fin = pd.concat([f, corr], axis=1)
            fin.columns = ['adg_score','correlation']
            fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
            nume = fin['score'].sum()
            deno = fin['correlation'].sum()
            final_score = avg_user + (nume/deno)
            score['user_id'].append(user)
            score['item_id'].append(item)
            score['Rating'].append(final_score)
      
        final_list.append(score)
    return final_list
```

```python
import pandas as pd
import numpy as np 
# numpy..numerical python deals with multidimensional array...matrices
import math
```

```python
Smartphones=pd.read_csv('smartphone.csv')
Ratings=pd.read_csv('ratings.csv')

Mean= Ratings.groupby(['user_id'], as_index = False, sort = False).mean().rename(columns = {'ratings': 'rating_mean'})[['user_id','rating_mean']]
Mean
```

```python
Ratings= pd.merge(Ratings, Mean, on ='user_id')
Ratings['rating_adjusted']=Ratings['ratings']-Ratings['rating_mean']
Ratings
```

```python
Ratings1= Ratings.fillna(0)
Ratings1
```

```python
Smartphones.head()
```

```python
check=pd.pivot_table(Ratings,index='user_id',columns='smartphone_id',values='ratings') #value with actual ratings
check
```

```python
final=pd.pivot_table(Ratings,index='user_id', columns='smartphone_id', values='rating_adjusted')
final
```

```python
final_user = final.apply(lambda row: row.fillna(row.mean()), axis=1)
#matrix with NaN value is a sparse matrix
final_user
```

```python
final_gadget = final.fillna(final.mean(axis=0))
final_gadget
```

```python

from sklearn.metrics.pairwise import cosine_similarity
b = cosine_similarity(final_user)
np.fill_diagonal(b, 0 )
similarity_with_user = pd.DataFrame(b,index=final.index)
similarity_with_user.columns=final.index
similarity_with_user
```

```python
cosine = cosine_similarity(final_gadget)
np.fill_diagonal(cosine, 0 )
similarity_with_gadget = pd.DataFrame(cosine,index=final_gadget.index)
similarity_with_gadget.columns=final_user.index
similarity_with_gadget.head()
```

```python
def find_n_neighbours(df,n):
    order = np.argsort(df.values, axis=1)[:, :n]
    df = df.apply(lambda x: pd.Series(x.sort_values(ascending=False)
           .iloc[:n].index, 
          index=['top{}'.format(i) for i in range(1, n+1)]), axis=1) #axis 0 for index 1 for columns
    return df
```

```python
# top 3 neighbours for each user
sim_user_3_u = find_n_neighbours(similarity_with_user,3)
sim_user_3_u.head()
```

```python
#lets now check our similarity result from the function...
def check_similarity(user1,user2):
    checked=Ratings[Ratings.user_id==user1].merge(Ratings[Ratings.user_id==user2],on ='id',how='inner')
    
    return checked.merge(Smartphones,on='id')
```

```python
Ratings = Ratings.astype({"smartphone_id": str})
gadget_user = Ratings.groupby(by = 'user_id')['smartphone_id'].apply(lambda x:','.join(x))
gadget_user
```

```python
#USER ITEM SCORE USING SCORE FUNCTION

def User_item_score2(user):
    gadget_seen_by_user=check.columns[check[check.index==user].notna().any()].tolist()
    a=sim_user_3_u[sim_user_3_u.index==user].values #similar users to a particular user
    b=a.squeeze().tolist()
    c=gadget_user[gadget_user.index.isin(b)]
    l=','.join(c.values)
    gadget_seen_by_similar_users=l.split(',')
    gadget_under_considiration=list(set(gadget_seen_by_similar_users)-set(list(map(str,gadget_seen_by_user))))
    gadget_under_considiration=list(map(int, gadget_under_considiration))
    score=[]
    for item in gadget_under_considiration:
        d=final_user.loc[:,item] #sabai uer le tyo item lai deko ratings
        e=d[d.index.isin(b)] #similar users le deko ratings extract gariyo
        f=e[e.notnull()]
        avg_user=Mean.loc[Mean['user_id']==user,'rating_mean'].values[0]
        index = f.index.values.squeeze().tolist() #neighbour users
        corr=similarity_with_user.loc[user,index]
        fin = pd.concat([f, corr], axis=1)
        fin.columns = ['adg_score','correlation']
        fin['score']=fin.apply(lambda x:x['adg_score'] * x['correlation'],axis=1)
        nume = fin['score'].sum()
        deno = fin['correlation'].sum()
        final_score = avg_user + (nume/deno)
        score.append(final_score)
    data = pd.DataFrame({'id':gadget_under_considiration,'score':score})
    top_recommendation=data.sort_values(by='score',ascending=False).head(4)
    gadget_name=top_recommendation.merge(Smartphones, how='inner', on='id')
    gadget_id=gadget_name.id.values.tolist()
    return(gadget_id)
```
